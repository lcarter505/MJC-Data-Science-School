**Comment**: Using value_counts will group and count how many times a unique key appears. The value_counts() return a pandas series, so you are accessing the 25th position in this series. However, if you check the value_counts series, you will see other months in 2025. Therefore, you are missing other data, and you need to sum up these ones too. One way to get this is by using a boolean mask and then summing up the values. True counts as 1 and False as 0. Thus, you can use this code for instance:

```python
print('We have ',(ecom['CC Exp Date'].apply(lambda x: x[3:]) == '25').sum(),
      ' clients that have credit cards that expires in 2025.')
```

```python
ecom['CC Exp Date'].value_counts()[25]
```

**Question 21**(Challenge) - How many people have the word Chief in their job title? (This is pretty tricky)
<br>
**Hint:**<br> 1) define a function to check the series of the attribute column;<br>2) Use the apply method with the lambda to call your defined function to get the sum. [apply method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)

```python
def chief_in_title(df, column, pattern):
  result = df[column].str.contains(pat = pattern).value_counts()
  return result
   
chief_in_title(df_sf, 'JobTitle', 'Chief')
```

```python
chief_in_title = lambda x: df_sf['JobTitle'].str.contains(pat = 'Chief').value_counts()
df_sf.apply(chief_in_title)['JobTitle']
   
```
