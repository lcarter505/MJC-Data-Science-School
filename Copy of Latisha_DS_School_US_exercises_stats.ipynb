**Questio 1)** For the dataset bellow, calculates the following properties:

data: 15, 4, 3, 8, 15, 22, 7, 9, 2, 3, 3, 12, 6

 - n
 - median
 - first quartile
 - third quartile
 - mean
 - mode


```python
series = np.array([15, 4, 3, 8, 15, 22, 7, 9, 2, 3, 3, 12, 6])
print('n:', len(series))
sorted_series = np.sort(series)
print(sorted_series)
print('median:', np.median(series))
print('first quartile:', np.quantile(series,0.25,interpolation='midpoint'))
print('third quartile:', np.quantile(series,0.75,interpolation='midpoint'))
print('mean:', np.mean(series))
print('mode:', stats.mode(series))
```

**Question 2)** For the dataset bellow, calculates the following properties:

data: 15, 4, 3, 8, 15, 22, 7, 9, 2, 3, 3, 12, 6

 - interquartile range
 - minimum
 - maximum
 - range
 - variance
 - standard deviation

```python
series = np.array([15, 4, 3, 8, 15, 22, 7, 9, 2, 3, 3, 12, 6])
iqr = np.quantile(series,0.75,interpolation='midpoint') - np.quantile(series,0.25,interpolation='midpoint')
print('interquartile range:', iqr)
print('minimum:', np.min(series)) 
print('maximum:', np.max(series)) 
range = (np.max(series) - np.min(series))
print('range:', range)
print('varianvce:', np.var(series))
print('variance check (var squared = std):', np.sqrt(np.var(series)))
print('standard deviation:', np.std(series))
```

**Question 3)** Create a function to calculate the mean of the following list. Then compare to numpy

List: 15, 4, 3, 8, 15, 22, 7, 9, 2, 3, 3, 12, 6

Mean formula:<br>

$$ \overline{x} = \frac{1}{n}(\sum_{i=1}^{n}x_{i}) = \frac{x_{1}+x_{2}+\cdots +x_{n}}{n} $$

```python
def mean_formula(list):
  mean = sum(list)/len(list)

  return mean
```

```python
list = [15, 4, 3, 8, 15, 22, 7, 9, 2, 3, 3, 12, 6]

mean_formula(list)
```

**Question 4)** (Optional - Challenge) Using the same data in the previous question, calculate the **median**<br>
**Hint**: <br>1) Sort the data; <br>2) Check the array size: is it even or odd? <br>3) If even, calculates the mean for the two center values; <br>4) Otherwise (odd), get the center value of the array.

```python
def median_formula(list):
  list.sort()
  middle = (len(list) - 1)/2

  if middle % 2 == 0:
    median = list[int(middle)]
  else:
    middle_1 = middle - .5
    middle_2 = middle + .5
    median = (middle_1 + middle_2)/2
    
  
  return list, median
```
